##### Explaining the Equations:
The derivative is effectively a measure of some rate of change. Here we have x which can be a scalar or a array. To approximate the derivative using continous values, the rate of change must be computed at each point on the function with respect to x. f(x) or y is what is being derived with respect. As such, we measure f(x) and difference it to a point behind (f(x-h)) or in front (f(x+h)) with respect the change in x = (x+h-x = h). The best solution is for h â†’ 0. We see this is the second order error term $-\frac{h^2}{2}fâ€³(x)$ where a lower h significantly reduces error (rate of convergence). 

Truncation error is measured in terms of error. Nth order derivatives have error of the order of $O(h^n)$

To reduce error, we combine back and forward differencing into central differencing. 


- *Center*: $f^â€²(x) â‰ˆ \frac{f(x+h)-f(x-h)}{2h}$
